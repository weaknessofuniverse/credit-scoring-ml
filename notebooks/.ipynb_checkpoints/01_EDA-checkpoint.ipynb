# 1. Load data

```python
df = pd.read_csv(train_path)
if 'Unnamed: 0' in df.columns:
    df = df.drop(columns=['Unnamed: 0'])

print('Shape:', df.shape)
df.head()
```

## Quick info and basic stats

```python
print("\nInfo:")
df.info()

print("\nDescribe (numeric):")
df.describe().T
```

## Missing values

```python
missing = df.isna().sum().sort_values(ascending=False)
print(missing)
```

```python
mi_missing_pct = df['MonthlyIncome'].isna().mean()
print(f"MonthlyIncome missing: {mi_missing_pct:.2%}")
```

## Target distribution

```python
print(df['SeriousDlqin2yrs'].value_counts(dropna=False))
print("Default rate:", df['SeriousDlqin2yrs'].mean())
```

## Strange / magic values in late-payment fields

```python
late_cols = [c for c in df.columns if 'NumberOfTime' in c or 'NumberOfTimes90DaysLate' in c]
for c in late_cols:
    print('\nColumn:', c)
print(df[c].value_counts().head(10))
```

Strategy note:
 - Replace these magic values with NaN (treat as missing) or a sentinel before imputation.
 - Another option: replace them with 0 if domain knowledge says so.


```python
# Replace 96 and 98 with np.nan in late-columns, we'll impute later.
magic_vals = {96,98}
for c in late_cols:
    df[c] = df[c].replace(list(magic_vals), np.nan)

# Check replacement
for c in late_cols:
    print(c, 'NA after replace:', df[c].isna().sum())
```

## Visualizations: distributions

```python
# Plot histograms for main numeric columns.
num_cols = [c for c in df.columns if df[c].dtype in [np.float64, np.int64] and c != 'SeriousDlqin2yrs']

fig, axes = plt.subplots(nrows=(len(num_cols) + 2) // 3, ncols=3, figsize=(18, 12))
axes = axes.flatten()

for i, col in enumerate(num_cols):
    axes[i].hist(df[col].dropna(), bins=50, color='steelblue', edgecolor='black')
    axes[i].set_title(col, fontsize=10)
    axes[i].tick_params(axis='x', rotation=45, labelsize=10)

for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.subplots_adjust(hspace=0.6, wspace=0.4)
plt.show()
```

## Log transform for skewed income

```python
# Create log1p(MonthlyIncome) for better distributional properties and plots.
df['log_MonthlyIncome'] = np.log1p(df['MonthlyIncome'])
plt.figure(); sns.histplot(df['log_MonthlyIncome'].dropna(), bins=60).set_title('log1p(MonthlyIncome)')
```

## Correlation matrix (pairwise)

```python
# Compute pairwise correlations on numeric columns (drop NaNs for pairwise computation).
plt.figure(figsize=(8,6))
corr = df[num_cols].corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation matrix')
```

## Target by binned features (example: age)

```python
# Create age bins and compute default rate per bin.
age_bins = [0,25,35,45,55,65,75,120]
df['age_bin'] = pd.cut(df['age'], bins=age_bins, right=False)
age_table = df.groupby('age_bin')['SeriousDlqin2yrs'].agg(['count','mean']).sort_values('mean', ascending=False)
age_table
```
